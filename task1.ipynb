```python
| Continent     | Internet Users (in millions) | %     |
| ------------- | ---------------------------- | ----- |
| Asia          | 2,700                        | 53%   |
| Europe        | 800                          | 15.7% |
| Africa        | 570                          | 11.2% |
| Latin America | 470                          | 9.3%  |
| North America | 400                          | 7.8%  |
| Oceania       | 30                           | 0.6%  |
```

```python
import matplotlib.pyplot as plt
continents = ['Asia', 'Europe', 'Africa', 'Latin America', 'North America', 'Oceania']
users = [2700, 800, 570, 470, 400, 30]
percentages = [53, 15.7, 11.2, 9.3, 7.8, 0.6]
colors = ['gold', 'skyblue', 'lightgreen', 'salmon', 'plum', 'lightgray']
plt.figure(figsize=(10, 6))
bars = plt.bar(continents, users, color=colors)
for bar, percent in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, f'{percent}%', ha='center', fontsize=11)
plt.title("Global Internet Users by Continent (2024)")
plt.ylabel("Users (Millions)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
